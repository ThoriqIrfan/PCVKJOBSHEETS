```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Image drive/Images/j.png')
plt.imshow(img)
```

```python
#infl- I mean dilation
kernel = np.ones((5,5), np.uint8)
dilation = cv2.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Initial Image')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation Results Image')
```

```python
#Erosion
kernel = np.ones((5,5), np.uint8)
erosion = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(erosion), plt.title('Erosion Results Image')
```

```python
#Dilation with OpenCV
ret, thresh = cv2.threshold(img, 127, 225, cv2.THRESH_BINARY)
kernel = np.ones((5,5),np.uint8)
dilasi = cv2.dilate(thresh,kernel,iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Initial Image')
plt.subplot(122),plt.imshow(dilasi),plt.title('Dilation Results Image')
```

```python
#Erosion with OpenCV
kernel = np.ones((5,5), np.uint8)
erosion = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(erosion), plt.title('Erosion Results Image')
```

```python
#Opening with OpenCV
kernel = np.ones((7,7), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(opening), plt.title('Opening Result Image')
```

```python
#Opening without OpenCV
element = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))
erosion_dst = cv2.erode(img, element)
opening_dst = cv2.dilate(erosion_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(opening), plt.title('Opening Result Image')
```

```python
#Closing with OpenCV
kernel = np.ones((7,7), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(opening), plt.title('Closing Result Image')
```

```python
#Closing without OpenCV
element = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))
erosion_dst = cv2.erode(img, element)
opening_dst = cv2.dilate(erosion_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(opening), plt.title('Closing Result Image')
```

```python
#Dilation Cross 3 and Cross 5

img = cv2.imread('/content/drive/MyDrive/Image drive/Images/j.png')
element = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
dilation = cv2.dilate(img, element, iterations=1)

print("SED = Cross 3")
plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(dilation), plt.title('Dilation Results Image')
plt.show()

element = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
dilation = cv2.dilate(img, element, iterations = 1)
print("SED = Cross 5")
plt.subplot(121),plt.imshow(img),plt.title('Initial Image')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation Results Image')
```

```python
#Dilation Circle 3 and Circle 5
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
dilation = cv2.dilate(img, element, iterations=1)
print("SED = circle 3")
plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(dilation), plt.title('Dilation Results Image')
plt.show()

element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
dilation = cv2.dilate(img, element, iterations=1)
print("SED = circle 5")
plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(dilation), plt.title('Dilation Results Image')
plt.show()
```

```python
#Erotion 3x5 rectangle
element = cv2.getStructuringElement(cv2.MORPH_RECT,(3,5))
Erotion = cv2.erode(img,element,iterations = 1)
print("SE = Rectangle 3x5")
plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(Erotion), plt.title('Erotion Result Image')
```

```python
#Erotion 5x7 rectangle
element = cv2.getStructuringElement(cv2.MORPH_RECT,(5,7))
Erotion = cv2.erode(img,element,iterations = 1)
print("SE = Rectangle 5x7")
plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(Erotion), plt.title('Erotion Result Image')
```

```python
#Dilation Vertical line 3
element = np.ones((3,1),np.uint8)
dilation = cv2.morphologyEx(img, cv2.MORPH_DILATE, element)
print("SE = Vertical line 3")
plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(dilation), plt.title('Dilation Result Image')
```
