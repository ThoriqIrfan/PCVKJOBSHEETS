```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image,cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
height = image_2.shape[0]
width = image_2.shape[1]
print("Image resolution: Height x Width = " , height," x ",width)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (width):
  image_3[int((height)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#vertical line
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (height):
  image_3[:,int((width)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#diagonals

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (height):
  image_3[y,int(y*(width/height))] =[255,255,255]
for y in range (width):
  image_3[int((height/width*y)),y] =[255,255,255]
  image_3[int((height/width*y)),(width-1-y)] =[255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#random short horizontal line
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_hori = cv.cvtColor(image, cv.COLOR_BGR2RGB)

line = 125

y_center = int(height / 2)
x_start = int((width - line) / 3)
x_end = x_start + line
for x in range(x_start, x_end):
    image_hori[y_center, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_hori))

cv2_imshow(final_frame)
```

```python
#box on white patch

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_with_box = cv.cvtColor(image, cv.COLOR_BGR2RGB)

x_box = 300
y_box = 130
width_box = 20
height_box = 20

for y in range(y_box, y_box + height_box):
    for x in range(x_box, x_box + width_box):
        image_with_box[y, x] = [0, 0, 255]

final_frame = cv.hconcat((image_2, image_with_box))
cv2_imshow(final_frame)
```
